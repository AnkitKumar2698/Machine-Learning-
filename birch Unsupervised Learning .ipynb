```python
import numpy as np
from matplotlib import pyplot as plt
# import sklearn.datasets.samples_generator import make_blobs
from sklearn.datasets import make_blobs
```

```python
X,clusters=make_blobs(n_samples=200,centers = 8, cluster_std = 0.75, random_state = 0)
```

```python
X
```

```python
plt.scatter(X[:,0],X[:,1],alpha=0.25)
```

```python
from sklearn.cluster import Birch
```

```python
brc=Birch(branching_factor=50,n_clusters=None,threshold=1.5)
```

```python
brc.fit(X)
```

```python
labels=brc.predict(X)
```

```python
plt.scatter(X[:,0],X[:,1],c=labels,cmap='rainbow',alpha=0.7)
```
